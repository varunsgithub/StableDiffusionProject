```python
from PIL import Image, ImageOps
from pillow_heif import register_heif_opener
import cv2
import numpy as np
import os
register_heif_opener()
```

```python
def preProcessData(img_path, output_path, img_size=(512,512)):
    with Image.open(img_path) as img:
        img = ImageOps.exif_transpose(img)
        img = img.convert("RGB")

        procImg = ImageOps.fit(img, img_size, method=Image.Resampling.LANCZOS)

        procImg.save(output_path, quality=100)
```

```python
inputDir = "/Users/varunsingh/Desktop/Course notes/CIS 5190/Final Project/Datasets/Consol"
outputDir = "/Users/varunsingh/Desktop/Course notes/CIS 5190/Final Project/Datasets/Resized and cleaned"

for filename in os.listdir(inputDir):
    if filename.lower().endswith(('.jpg')):
        input_path = os.path.join(inputDir, filename)
        output_path = os.path.join(outputDir, filename)
        preProcessData(input_path, output_path)
```

```python
# Canny image outline generation:
def cannyProcess(ipath, opath):
    img = cv2.imread(ipath)
    # Grayscale mask for ML model.
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # The brightness metric to test if it is day or night
    brightness = np.mean(gray)

    if (brightness < 70):
        #Its likely a night image
        low, high = 50, 100
        blurred = cv2.GaussianBlur(gray, (5,5), 0)
    else:
        low, high = 100, 260
        # We blur to ensure there are no jagged entries.
        blurred = cv2.GaussianBlur(gray, (3,3), 0)
    
    # edges = cv2.Canny(blurred, 150, 300)
    canny_map = cv2.Canny(blurred, low, high)

    cv2.imwrite(opath, canny_map, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
```

```python
inputDir = "/Users/varunsingh/Desktop/Course notes/CIS 5190/Final Project/Datasets/Resized and cleaned"
outputDir = "/Users/varunsingh/Desktop/Course notes/CIS 5190/Final Project/Datasets/CannyMap"

for filename in os.listdir(inputDir):
    if filename.lower().endswith(('.jpg')):
        input_path = os.path.join(inputDir, filename)
        output_path = os.path.join(outputDir, filename)
        cannyProcess(input_path, output_path)
```
